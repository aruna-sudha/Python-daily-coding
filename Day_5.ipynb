```python
# Project Title: Student Grades & Hobbies Tracker
 # Step 1: Create a dictionary for student grade
student_grades = {}



```

```python
# Step 2: Add or update grades for at least 3 students
print("Enter grades for 3 students:")
for i in range(3):
    name = input(f"Enter student {i+1} name: ")
    grade = input(f"Enter {name}'s grade: ")
    student_grades[name] = grade   # add/update directly

print("\nStudent Grades Dictionary:")
print(student_grades)

```

```python

# Step 3: Create a set of hobbies


hobbies = set()

print("\nEnter 3 hobbies:")
for i in range(3):
    hobby = input(f"Enter hobby {i+1}: ")
    hobbies.add(hobby)   # add to set

print("\nHobbies Set:")
print(hobbies)

```

```python

# Step 4: Convert set to frozenset
frozen_hobbies = frozenset(hobbies)

print("\nHobbies Frozenset (Immutable):")
print(frozen_hobbies)

```
