```python
# Project Title: Contact Directory
# Step 1: Create a 2D list with sample contacts
```

```python
contacts = [
    ["Ravi", "9876543210", "ravi@example.com"],
    ["Meera", "9123456780", "meera@example.com"],
    ["Sam", "9988776655", "sam@example.com"],
    ["Anu", "9001122334", "anu@example.com"],
    ["Raj", "9112233445", "raj@example.com"]
]

print(contacts)
```

```python

# Step 2: Allow user to add a new contact

```

```python
name = input("Enter contact name: ")
phone = int(input("Enter phone number: "))
email = input("Enter email address: ")
contacts.append([name, phone, email])

print(contacts)
print(f"\nContact {name} added successfully!")


```

```python

# Step 3: Display all contacts sorted alphabetically by name

contacts.sort()
print(contacts)
```

```python
# Step 4: Search contacts by name (partial match)
search_name = input("\nEnter name to search: ")
print(f"\nSearch Results for '{search_name}':")
found = False
for c in contacts:
    if search_name.lower() in c[0].lower():
        print("Name:", c[0], "| Phone:", c[1], "| Email:", c[2])
        found = True
if not found:
    print("No matching contacts found.")




```

```python

# Step 5: Display only the first 3 contacts after sorting

```

```python
print("\nTop 3 Contacts (Alphabetical Order):")
for c in contacts[:3]:
    print("Name:", c[0], "| Phone:", c[1], "| Email:", c[2])

```
